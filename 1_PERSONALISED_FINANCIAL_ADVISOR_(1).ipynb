### 3.8 Adding columns for "Total Expenditure" and "Average"

```python
df['total expenditure'] = df['housing'] + df['food'] + df['transport']+df['bills'] + df['personal_needs'] + df['debt_repayment']+df['family_needs'] + df['health_insurance']+df['entertainment_leisure']
df['average'] = df['total expenditure']/10
df.head()
```

Shows numberb of people with an expenditure above 80000

```python
reading_full = df[df['total expenditure'] >=80000].count()


print(f'Number of people with a monthly expenditure above 80000: {reading_full}')
```

shows number of people with an expenditure below 80000

```python
reading_less=df[df['total expenditure'] <=80000].count()


print(f'Number of people with a monthly expenditure below 80000: {reading_less}')
```

#####  Insights

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? )

```python
gender_group = df.groupby('gender')[['housing', 'family_needs', 'debt_repayment', "average"]].mean()
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total expenditure','Average']


female_scores = [gender_group['average'][0], gender_group['family_needs'][0]]
male_scores = [gender_group['average'][1], gender_group['family_needs'][1]]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')

plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total expenditure v/s family_needs expenditure of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights
- On an average males spend more on family needs.
- whereas females spend more overally.
